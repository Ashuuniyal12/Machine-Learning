## since we have 23 column and we dont want to have all of them for our model we cam remove unwanted ones and keep the important.

```python
# gener id keywords title overview cast crew
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
         L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(3)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(3)
```

```python
movies['cast'][0]
```

```python
#since we have a lot of dictionary of cast we can
#take only top 3 and convert it into a list
```

```python
def convert3(text):
    L = []
    counter =0
    for i in ast.literal_eval(text):
        if counter != 3:
             L.append(i['name'])
             counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
# in this part of dict we have a lot of key value pair but we need only job field defined as director 
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break;
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#now we have all list column except overview so convert overview as a list and concat all the columns as a single one named tags
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
#sam wermington
#sam Mendis they both can create difficulty in tags so concat them if there a space btw them
```

```python
def remove_space(L):
    L1  = []
    for i in L:
        L1.append(i.replace(" ", ""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
movies = movies.drop(columns= ['overview','genres','keywords','cast','crew'])
```

```python
movies.head()
```

```python
#conver the tags column data from list to a string data type for easy evaluation
```

```python
movies['tags'] = movies['tags'].apply(lambda x : " ".join(x))
```

```python
#convert all strings to lower case 
```

```python
movies['tags'] =  movies['tags'].apply(lambda x : x.lower())
```

```python
movies.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vector = cv.fit_transform(movies['tags']).toarray()
```

```python
vector.shape
```

```python
vector
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movieName): 
    movie_index = movies[movies['title']==movieName].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse= True,key =lambda x:x[1])
    
    for index in movies_list:
        print(movies.iloc[index[0]].title)
```

```python
recommend('U.F.O.')
```
